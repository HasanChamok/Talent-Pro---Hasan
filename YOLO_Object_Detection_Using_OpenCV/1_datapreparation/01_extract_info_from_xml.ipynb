```python
import os
from glob import glob
import pandas as pd
from functools import reduce
from xml.etree import ElementTree as et
```

```python
#load xml files and load it in a list
xmlfiles = glob('./data_images/*.xml')  #*.xml will run all the xml file extention
xmlfiles
```

```python
#replace \\ with /
replace_text = lambda x: x.replace('\\','/')
xmlfiles = list(map(replace_text,xmlfiles))
xmlfiles
```

```python
#step-2: read xml files
#from each xml file we need to extract 
#filename, size(width,height), object(name,xmin,xmax,ymin,ymax)
tree = et.parse('./data_images/000001.xml')
root = tree.getroot()
#Now we can extract all the information from the xml file
print(root)
#extract filename [For single xml file]
image_name = root.find('filename').text
#width and height of the image
width = root.find('size').find('width').text
height = root.find('size').find('height').text
print("Image name : " + image_name + " Width : " + width + " Height : " + height)

#finding the object form the xml files
objs = root.findall('object')
print(objs)
obj = objs[0] 
name = obj.find('name').text
bndbox = obj.find('bndbox')
xmin = bndbox.find('xmin').text
xmax = bndbox.find('xmax').text
ymin = bndbox.find('ymin').text
ymax = bndbox.find('ymax').text
print([name , xmin , xmax , ymin , ymax])
```

```python
#For multiple xml file we have to run it into a loop
image_name = root.find('filename').text
#width and height of the image
width = root.find('size').find('width').text
height = root.find('size').find('height').text
objs = root.findall('object')
parser=[]
for obj in objs:
    name = obj.find('name').text
    bndbox = obj.find('bndbox')
    xmin = bndbox.find('xmin').text
    xmax = bndbox.find('xmax').text
    ymin = bndbox.find('ymin').text
    ymax = bndbox.find('ymax').text
    parser.append([image_name, width, height, name , xmin , xmax , ymin , ymax])

print(parser)
```

```python
#For multiple xml file we have to run it into a loop
def extract_text(filename):
    tree = et.parse(filename)
    root = tree.getroot()
    image_name = root.find('filename').text
    #width and height of the image
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text
    objs = root.findall('object')
    parser=[]
    for obj in objs:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text
        parser.append([image_name, width, height, name , xmin , xmax , ymin , ymax])

    return parser
```

```python
parser_all = list(map(extract_text,xmlfiles))
print(parser_all)
```

```python
#reducing the data into 1d list
data = reduce(lambda x , y : x + y, parser_all)
print(data)
```

```python
#converting list into dataframe to train and test model
df = pd.DataFrame(data,columns=['filename','width','height','name','xmin','xmax','ymin','ymax'])
df.head()
```

```python
#type conversion
cols = ['width','height','xmin','xmax','ymin','ymax']
df[cols] = df[cols].astype(int)
df.info()
```

```python
df['center_x'] = ((df['xmin']+df['xmax'])/2)/df['width']
df['center_y'] = ((df['ymin']+df['ymax'])/2)/df['height']
df['w'] = (df['xmax']-df['xmin'])/df['width']
df['h'] = (df['ymax']-df['ymin'])/df['height']
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
unique_labels = df['name'].unique()
def label_encoding(x):
    # Assuming img_df is your DataFrame and 'filename' is the column containing labels
    
    
    # Initialize LabelEncoder
    label_encoder = LabelEncoder()
    
    # Fit and transform the unique labels to numerical values
    encoded_labels = label_encoder.fit_transform(unique_labels)
     # To map numerical labels back to original labels
    decoded_labels = label_encoder.inverse_transform(encoded_labels)
    # Combine labels and encoded labels into a dictionary
    labels = dict(zip(decoded_labels, encoded_labels))
    
    # # Print the encoded labels
    # print("Encoded Labels:", encoded_labels)
    # print("Decoded Labels:", decoded_labels)
    # # Print the dictionary
    # print("Label Encoding Dictionary:", labels)

    return labels[x]
```

```python
#label encoding
def label_encoding(x):
    labels = {'person':0,'car':1,'chair':2,'dog':3,'boat':4,'bottle':5,'bird':6,'bicycle':7,'horse':8,'diningtable':9,'motorbike':10,'pottedplant':11,'aeroplane':12,'train':13,'sofa':14,'cat':15,'sheep':16,'tvmonitor':17,'bus':18,'cow':19}
    return labels[x]
```

```python
train_df['id'] = train_df['name'].apply(label_encoding)
test_df['id'] = test_df['name'].apply(label_encoding)
```

```python


#SAVE image and Label in text

import os
from shutil import move
```

```python
train_folder = 'data_images/train'
test_folder = 'data_images/test'

os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python
 cols = ['filename','id','center_x','center_y','w','h']
groupby_obj_train = train_df[cols].groupby('filename')
groupby_obj_test = test_df[cols].groupby('filename')
```

```python
# groupby_obj_train.get_group('000001.jpg').set_index('filename').to_csv('sample.txt',index=False,header=False)
def save_data(filename,folder_path,group_obj):
    src = os.path.join('data_images',filename)
    dst = os.path.join(folder_path,filename)
    move(src,dst)

    text_filename = os.path.join(folder_path,os.path.splitext(filename)[0]+'.txt')
    group_obj.get_group(filename).set_index('filename').to_csv(text_filename,sep=' ',index=False,header=False)
    
```
